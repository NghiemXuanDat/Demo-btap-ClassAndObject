```python
class Ingredient:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity
```

```python
from tkinter import *
from tkinter import messagebox
# Tạo ra cửa sổ ứng dụng
window = Tk()

# Set up thông tin cho window 
window.title("Ingredients Management System")
window.geometry("350x150")

# Tạo list để lưu trữ toàn bộ các object trong list
list_ingredients = []

# Mô tả các function (add/search/update/delete/reset)
def add_ingredient():
    # lấy giá trị từ entry
    name_value = name.get()
    quantity_value = int(quantity.get())

    new_ingredient = Ingredient(name_value, quantity_value)

    # Thêm vào list
    list_ingredients.append(new_ingredient)

    # List comprehension
    print([item.name for item in list_ingredients])

    # Hiển thị thông báo
    messagebox.showinfo("Add Ingredient", f"{name_value} is added successfully")

    # Xóa giá trị ở entry
    name.delete(0, END)
    quantity.delete(0, END)

def search_ingredient():
    search_value = search_name.get()
    is_exist = False
    for ingredient in list_ingredients:
        if ingredient.name == search_value:
            is_exist = True
            messagebox.showinfo("Search Ingredient", f"Name: {ingredient.name}, quantity: {ingredient.quantity}")
            break

    if is_exist == False:
        messagebox.showinfo("Search Ingredient", f"{search_value} is not found")

def delete_ingredient():
    search_value = search_name.get()
    is_exist = False
    for ingredient in list_ingredients:
        if ingredient.name == search_value:
            is_exist = True
            messagebox.showinfo("Delete Ingredient", f"{ingredient.name} is deleted successfully")
            break

    if is_exist == False:
        messagebox.showinfo("Delete Ingredient", f"{search_value} is not found")

def update_ingredient():
    name_value = name.get()
    quantity_value = int(quantity.get())

    is_exist = False
    for ingredient in list_ingredients:
        if ingredient.name == name_value:
            is_exist = True
            ingredient.quantity = quantity_value 
            messagebox.showinfo("Update Ingredient", f"{ingredient.name} is updated successfully")
            break

    if is_exist == False:
        messagebox.showinfo("Update Ingredient", f"{name_value} is not found")


def reset():
    name.delete(0, END)
    quantity.delete(0, END)
    search_name.delete(0, END)

# set up các widget

# tạo label (2)
name_lbl = Label(window, text="Name:") 
quantity_lbl = Label(window, text="Quantity:")

# tạo entry (3)
name = Entry(window)
quantity = Entry(window)
search_name = Entry(window)

# tạo buton (5)
add_btn = Button(window, text="Add", command=add_ingredient)
search_btn = Button(window, text="Search", command=search_ingredient)
update_btn = Button(window, text="Update", command=update_ingredient)
delete_btn = Button(window, text="Delete", command=delete_ingredient)
reset_btn = Button(window, text="Reset", command=reset)

# Dán các widget lên window
name_lbl.grid(row=0, column=0)
quantity_lbl.grid(row=1, column=0)
name.grid(row=0, column=1)
quantity.grid(row=1, column=1)
search_name.grid(row=3, column=0)
add_btn.grid(row=2, column=0)
search_btn.grid(row=3, column=1)
update_btn.grid(row=3, column=2)
delete_btn.grid(row=2, column=1)
reset_btn.grid(row=2, column=2)

# Chạy ứng dụng
window.mainloop()

```
